## Inference

### Setting

```python
import os
import sys

from scripts.data.newmap_prep import build_new_map_arrays
from scripts.inference.rhythm_calc_aff import predict_notes
from scripts.inference.final_aff import build_aff_from_predictions

NB_DIR = os.path.abspath(os.getcwd())
ROOT_DIR = NB_DIR
ARCAEA_DIR = NB_DIR

if ROOT_DIR not in sys.path:
    sys.path.append(ROOT_DIR)

ACCEL_DEVICE = "auto"

USER_SKELETON_AFF = "inputs/3.aff"
USER_AUDIO_NAME = "inputs/base.ogg"
NOTE_DENSITY = 0.5 #[0,1]
MAX_FINGER_COUNT = 2
```

### Run Classifier

```python
map_ticks, map_timestamps, map_wav, map_extra = build_new_map_arrays(
    USER_SKELETON_AFF,
    audio_name=USER_AUDIO_NAME,
)

user_rhythm_npz = predict_notes(
    note_density=NOTE_DENSITY,
    device=ACCEL_DEVICE,
    aff_path=USER_SKELETON_AFF,
    ticks_in=map_ticks,
    timestamps_in=map_timestamps,
    wav_in=map_wav,
    extra_in=map_extra,
    max_finger_count=MAX_FINGER_COUNT,
)

NOTE_DENSITY, user_rhythm_npz
```

### Run GAN

```python
generated_aff = build_aff_from_predictions(
    USER_SKELETON_AFF,
    user_rhythm_npz,
)
generated_aff
```
